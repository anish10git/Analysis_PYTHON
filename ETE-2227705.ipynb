#VISUALIZATION

```python
import matplotlib.pyplot as plt
# Count the number of occurrences of each job
job_counts = final_df['Department'].value_counts()
# Create a bar plot
plt.bar(job_counts.index, job_counts.values)
# Set the x-axis label
plt.xlabel('department')
# Set the y-axis label
plt.ylabel('Number of People')
# Set the plot title
plt.title('attrition in deparements ')
# Show the plot
plt.show()
```

  


---


After comparing attrition rates across different departments, it was found that the Research & Development department has the highest attrition rate. However, it should be noted that the attrition rate in the Sales department is also significantly high, with over 60% of the rate seen in the Research & Development department.

As a result, it can be concluded that our focus should be on addressing attrition in both the Research & Development and Sales departments, as they both carry significant weight in terms of employee turnover.

This information is derived from a pie chart that visually represents the proportion of attrition across different job roles in the department.

Based on this finding, it can be concluded that the company should focus on addressing attrition among Research Scientists and Laboratory Technicians. Implementing measures such as conducting surveys to identify the reasons for attrition and providing opportunities for career growth and development could be some of the ways to improve employee retention in these job roles.

```python
import seaborn as sns

# Create a violin plot
sns.violinplot(x=final_df['PercentSalaryHike'],palette='Purples_r')
# Set the x-axis label
plt.xlabel('PercentSalaryHike')
# Set the plot title
plt.title('Distribution of Salary hike percentage')
# Show the plot
plt.show()
```

```python
# Select the five columns you want to use for the radar chart
columns = ['JobInvolvement', 'JobSatisfaction', 'WorkLifeBalance', 'PerformanceRating', 'RelationshipSatisfaction']
# Calculate the mean value of each column
means = final_df[columns].mean().tolist()
# Calculate the angle for each axis in the plot
angles = [i/float(len(columns))*(2*3.14159) for i in range(len(columns)+1)]
angles = angles[:-1]
# Create a radar chart
ax = plt.subplot(111, polar=True)
ax.set_theta_offset(3.14159 / 2)
ax.set_theta_direction(-1)
plt.xticks(angles, columns)
ax.set_rlabel_position(0)
# Plot the data
plt.plot(angles, means, 'o-', linewidth=2, color='purple')
plt.fill(angles, means, alpha=0.25, color='pink')
# Set the plot title
plt.title('Radar Chart rating scale outoff 4')
# Show the plot
plt.show()
```


---
By using the radar chart, we were able to compare the various ratings provided by employees, which ranged from 1 to 4, with higher ratings indicating higher levels of satisfaction. The analysis revealed that employees rated their job satisfaction highly, with a mean rating of more than 2.5. This suggests that there may not be any issues related to job description or work-life balance that are causing attrition.

However, despite the good ratings, employees are still leaving the company. This indicates that there may be other factors contributing to the high attrition rate, which need to be identified and addressed.

Therefore, further investigation is required to identify these other factors and develop strategies to retain employees and reduce attrition rates.

Therefore, it is recommended that the company reconsider their compensation structure and evaluate the possibility of offering higher salaries or monthly rates that are more competitive with industry standards. This can help to improve employee retention and reduce attrition rates in the long run.

```python
# Select the columns you want to use for the stacked bar chart
columns = ['MaritalStatus','Gender']
# Group the data by the two columns and count the occurrences of each combination
grouped = final_df.groupby(columns).size().unstack()
# Create a stacked bar chart
grouped.plot(kind='bar', stacked=True)
# Set the plot title
plt.title('Stacked Bar Chart')
# Show the plot
plt.show()
```



---
The graph shows that single employees, both male and female, have a higher attrition rate compared to employees with other marital statuses. This suggests that there may be a correlation between being single and leaving the company.

One possible explanation for this trend is that single employees may prioritize higher salaries in order to establish a solid financial foundation before getting married. As a result, they may be more likely to switch companies in search of better salary packages.

Therefore, it is recommended that the company review its compensation structure and consider offering more competitive salaries to retain single employees. Additionally, other factors that may be contributing to the higher attrition rate among single employees, such as career advancement opportunities, work-life balance, and job satisfaction, should also be examined and addressed.

```python
# Select the columns for the box plot
columns = ['MonthlyIncome', 'JobRole']
# Create a box plot with a larger figure size
fig, ax = plt.subplots(figsize=(17,8))
final_df.boxplot(column='MonthlyIncome', by='JobRole', grid=False, ax=ax)
# Set the title of the plot
plt.title('Box Plot of income and Job Role')
# Show the plot
plt.show()
```

        

---
The box plot analysis reveals that job roles with lower salaries, such as laboratory technicians, research scientists, and sales representatives, have a smaller range between the highest and lowest salaries. This suggests that there is not much variation in salary within these job roles.

Furthermore, the green line on the box plot, which represents the average salary, is positioned closer to the lower end of the salary range for these job roles. This indicates that employees in these job roles may be receiving lower salaries compared to other job roles.

Based on this analysis, it can be concluded that salary is a major factor contributing to the high attrition rate. Employees in job roles with lower salaries may not be satisfied with their current salary and may be seeking better opportunities elsewhere.

Therefore, the company should evaluate the compensation structure for job roles with lower salaries and consider offering more competitive salaries to retain employees in these positions. Additionally, the company should explore other factors that may be contributing to the high attrition rate and address them accordingly.

```python
import seaborn as sns
# Select the columns you want to use for the heatmap
columns = ['JobRole','OverTime']
# Group the data by the two columns and count the occurrences of each combination
grouped = final_df.groupby(columns).size().unstack()
# Create the heatmap using the grouped data
sns.heatmap(grouped, cmap='YlOrRd', annot=True, fmt='d')
# Set the plot title
plt.title('Heatmap')
# Show the plot
plt.show()
```


---
The heat map analysis reveals that employees in certain job roles, such as laboratory technician, research scientist, sales executive, and sales representative, are working overtime more frequently compared to other job roles. The darker color in these cells indicates a higher count of employees working overtime.

This finding is consistent with the previous analysis, which showed that employees in these job roles are receiving lower salaries and may not be satisfied with their compensation. Additionally, the fact that these employees are working overtime suggests that they may be feeling overworked and may not have a healthy work-life balance, which could contribute to job dissatisfaction and ultimately lead to attrition.

Therefore, it is recommended that the company evaluates the workload and work-life balance for employees in these job roles and consider offering more competitive salaries, as well as providing opportunities for career advancement and professional development to increase job satisfaction and retention.
